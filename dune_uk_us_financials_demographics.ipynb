```python
# dune_dataset_template.py
import csv
import datetime
from pathlib import Path

# ---------------------------------------------
# CONFIGURABLE PARAMETERS
# ---------------------------------------------
# Choose your time range here. Examples:
# 1) Past year: start_date = today - 1 year, end_date = today
# 2) Past 5 years: start_date = today - 5 years, end_date = today
# 3) Custom: set explicit start_date and end_date
#
# For demo/template, we'll default to past 12 months.
today = datetime.date.today()
start_date = today.replace(day=1) - datetime.timedelta(days=365)
start_date = start_date.replace(day=1)  # first day of that month
end_date = today.replace(day=1)  # first day of current month

# If you want a custom range, uncomment and set:
# start_date = datetime.date(2021, 1, 1)
# end_date = datetime.date(2025, 12, 1)

# Path to output directory
output_dir = Path("dune_data")
output_dir.mkdir(exist_ok=True)

# ---------------------------------------------
# HELPERS
# ---------------------------------------------
def month_range(start, end):
    """
    Yield first day of each month from start (inclusive) to end (exclusive).
    """
    cur = start
    while cur < end:
        yield cur
        # advance to next month
        year = cur.year + (cur.month // 12)
        month = cur.month % 12 + 1
        cur = datetime.date(year, month, 1)

def write_csv(path, headers, rows):
    with path.open("w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(rows)
    print(f"Written {path} with {len(rows)} data rows.")

# ---------------------------------------------
# TEMPLATE STRUCTURE
# ---------------------------------------------
#
# A) Brand‑wide, product category, regional, store‑level, monthly tables
#
# Columns explanation:
# - Region: UK or US
# - Month: YYYY-MM
# - Level: Brand, Category, Region, Store
# - Level_ID / Name: depending on Level; e.g., for Brand use 'Dune'; for Category use category name; for Region use 'London', 'NYC', etc.; for Store use store code or name
# - Sales: total revenue in local currency
# - Cost: total cost of goods sold or direct cost
# - Profit: Sales - Cost
# - Profit_Margin: Profit / Sales, decimal or percent
# - Units_Sold: optional, units
# - Notes: optional text
#
# Example rows are placeholders; you must replace them with real data.

headers = [
    "Region",
    "Month",          # YYYY-MM
    "Level",          # Brand, Category, Region, Store
    "Level_ID",       # e.g. "Dune" or category name or region name or store code
    "Level_Name",     # human-readable
    "Sales",          # numerical, local currency
    "Cost",           # numerical, local currency
    "Profit",         # computed or input
    "Profit_Margin",  # Profit / Sales, e.g. 0.12 for 12%
    "Units_Sold",     # optional
    "Notes",          # optional
]

# B) Demographics tables: dimensional breakdown per month
#
# Demographic fields: Age, Gender, Income, Ethnicity
# You can record percentages or counts (choose one consistently)
# Example columns:
demo_headers = [
    "Region",
    "Month",          # YYYY-MM
    "Level",          # Brand, Category, Region, Store
    "Level_ID",
    "Level_Name",
    "Age_Group",      # e.g. "18-24"
    "Gender",         # e.g. "Male", "Female", "Other"
    "Income_Bracket", # e.g. "30-50k GBP" or "50-75k USD"
    "Ethnicity",      # e.g. "White", "Black", "Asian", "Hispanic", "Other"
    "Count",          # or "Percentage"
    "Notes",          # optional
]

# ---------------------------------------------
# CREATE MONTHLY ROWS FOR UK AND US
# ---------------------------------------------
def generate_empty_rows(region_code):
    rows = []
    for m in month_range(start_date, end_date):
        month_str = m.strftime("%Y-%m")
        # Brand-wide placeholder
        rows.append([region_code, month_str, "Brand", "Dune", "Dune", "", "", "", "", ""])
        # Add placeholders for categories
        example_categories = ["Footwear", "Bags", "Accessories"]
        for cat in example_categories:
            rows.append([region_code, month_str, "Category", cat, cat, "", "", "", "", ""])
        # Add example regions / zones
        example_regions = []
        if region_code == "UK":
            example_regions = ["London", "Manchester", "Birmingham"]
        else:
            example_regions = ["New York", "Los Angeles", "Chicago"]
        for reg in example_regions:
            rows.append([region_code, month_str, "Region", reg, reg, "", "", "", "", ""])
        # Add example stores; customize store codes to your actual data
        example_stores = []
        if region_code == "UK":
            example_stores = ["UK-001", "UK-002", "UK-003"]
        else:
            example_stores = ["US-001", "US-002", "US-003"]
        for store in example_stores:
            rows.append([region_code, month_str, "Store", store, store, "", "", "", "", ""])
    return rows

uk_rows = generate_empty_rows("UK")
us_rows = generate_empty_rows("US")

# Demographic rows: start with empty placeholders; you can expand or generate proactively per combination
def generate_empty_demo_rows(region_code):
    rows = []
    for m in month_range(start_date, end_date):
        month_str = m.strftime("%Y-%m")
        # Example: one placeholder per month; in practice add one per Level + demo combination
        rows.append([region_code, month_str, "Brand", "Dune", "Dune", "", "", "", "", "", ""])
        # Add more rows as you collect demographic distributions for store, category, region etc.
    return rows

uk_demo_rows = generate_empty_demo_rows("UK")
us_demo_rows = generate_empty_demo_rows("US")

# ---------------------------------------------
# WRITE CSV FILES
# ---------------------------------------------
write_csv(output_dir / "dune_uk_monthly.csv", headers, uk_rows)
write_csv(output_dir / "dune_us_monthly.csv", headers, us_rows)
write_csv(output_dir / "dune_uk_monthly_demographics.csv", demo_headers, uk_demo_rows)
write_csv(output_dir / "dune_us_monthly_demographics.csv", demo_headers, us_demo_rows)

print("Template CSV files created. Fill in the numeric values and demographic counts/percentages from your source data.")
```
