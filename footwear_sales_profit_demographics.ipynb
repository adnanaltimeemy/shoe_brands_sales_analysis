```python
import pandas as pd
from datetime import date

# ---------------------------------------------------------------------
# Helper to create a standardized record
# ---------------------------------------------------------------------
def make_record(
    brand,
    region,
    period_start,
    period_end,
    currency,
    sales=None,                 # sales revenue in currency
    profit_margin=None,         # e.g., decimal 0.055 for 5.5%
    notes="",                   # source notes, assumptions
    # Demographics: dictionaries or strings; adjust as needed
    demo_age_distribution=None, # e.g., {"18-24":0.2, "25-34":0.35, ...}
    demo_gender_distribution=None,
    demo_income_distribution=None,
    demo_other=None
):
    return {
        "brand": brand,
        "region": region,
        "period_start": period_start,
        "period_end": period_end,
        "currency": currency,
        "sales": sales,
        "profit_margin": profit_margin,
        "notes": notes,
        "demo_age_distribution": demo_age_distribution,
        "demo_gender_distribution": demo_gender_distribution,
        "demo_income_distribution": demo_income_distribution,
        "demo_other": demo_other
    }

# ---------------------------------------------------------------------
# Example records with public figures where available
# ---------------------------------------------------------------------

# 1) Schuh, UK – FY ~early 2024 period, public turnover and profit data
# Source: Schuh UK 2023 results reporting turnover and profit. 
# Turnover £380.8M; profit before tax £21M. :contentReference[oaicite:0]{index=0}
# Profit margin approx = 21 / 380.8 ≈ 0.055 = 5.5%
schuh_uk_record = make_record(
    brand="Schuh",
    region="UK",
    period_start="2023-02-01",  # approximate start of the fiscal period referenced
    period_end="2024-01-31",    # approximate end
    currency="GBP",
    sales=380.8e6,              # £380.8M
    profit_margin=21e6 / 380.8e6,
    notes="Public report of turnover and pre-tax profit for period ending early Feb 2024. Footfall and LFL e-commerce growth noted. :contentReference[oaicite:1]{index=1}",
    # Demographics not publicly provided; fill later
    demo_age_distribution=None,
    demo_gender_distribution=None,
    demo_income_distribution=None,
    demo_other=None
)

# 2) Skechers, US or global – 2024 full year public result
# Public source: Skechers FY 2024 sales $8.97B; gross margin 53.3%. :contentReference[oaicite:2]{index=2} :contentReference[oaicite:3]{index=3}
# This is global, but you can note region or restrict later. 
skechers_global_record = make_record(
    brand="Skechers",
    region="Global",  # adjust or split to US when better data available
    period_start="2024-01-01",
    period_end="2024-12-31",
    currency="USD",
    sales=8.97e9,
    profit_margin=None,  # profit margin not directly given in that release; could compute if net profit known
    notes="FY 2024 global sales reported; gross margin 53.3% noted for Q4. Adjust for region split if available.", 
    # Add gross margin in a custom field if desired, or compute externally
    demo_age_distribution=None,
    demo_gender_distribution=None,
    demo_income_distribution=None,
    demo_other=None
)

# 3) Birkenstock, global – FY 2024 public result
# Source: Revenue €1.8B; gross profit margin 58.8%. :contentReference[oaicite:4]{index=4}
birkenstock_global_record = make_record(
    brand="Birkenstock",
    region="Global",
    period_start="2023-10-01",  # FY ended Sept 30 2024
    period_end="2024-09-30",
    currency="EUR",
    sales=1.8e9,
    profit_margin=None,  # use gross margin separately if desired
    notes="Fiscal 2024 revenue reported; gross profit margin 58.8%. :contentReference[oaicite:5]{index=5}",
    demo_age_distribution=None,
    demo_gender_distribution=None,
    demo_income_distribution=None,
    demo_other=None
)


# ---------------------------------------------------------------------
# Create dataframes for UK and US CSVs; initially assignment by region
# ---------------------------------------------------------------------

# UK CSV: include only records for UK or those intended for UK analysis
uk_records = [schuh_uk_record]  # add more UK-specific entries or split global entries later

# US CSV: include US records; currently Skechers/Birkenstock global as placeholders
us_records = [skechers_global_record, birkenstock_global_record]  # add US-specific entries later

# Convert to dataframes
df_uk = pd.DataFrame(uk_records)
df_us = pd.DataFrame(us_records)

# ---------------------------------------------------------------------
# Output CSVs
# ---------------------------------------------------------------------
output_dir = "."  # change to folder you want
uk_csv_path = f"{output_dir}/footwear_data_UK.csv"
us_csv_path = f"{output_dir}/footwear_data_US.csv"

df_uk.to_csv(uk_csv_path, index=False)
df_us.to_csv(us_csv_path, index=False)

print(f"UK CSV written to: {uk_csv_path}")
print(f"US CSV written to: {us_csv_path}")
```
