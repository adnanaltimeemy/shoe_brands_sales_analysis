```python
import pandas as pd
from pathlib import Path

# ============================================================
# 1) SALES / PROFIT DATA
# ============================================================

columns_sales = [
    "year",
    "region",
    "channel",
    "revenue_gbp",
    "gross_profit_gbp",
    "operating_profit_gbp",
    "pre_tax_profit_gbp",
    "net_profit_gbp",
    "notes"
]

sales_rows = []

# -------------------------
# 2023 Global (Combined)
# -------------------------
sales_rows.append({
    "year": 2023,
    "region": "Global",
    "channel": "Combined",
    "revenue_gbp": 994.5,
    "gross_profit_gbp": 481.1,
    "operating_profit_gbp": -20.3,
    "pre_tax_profit_gbp": -39.8,
    "net_profit_gbp": -32.1,
    "notes": "Public filing; FashionNetwork Jun 3 2024."
})

# -------------------------
# 2023 Regional Split
# -------------------------
regional_2023_data = [
    ("UK/ROI", 398.3),
    ("Americas", 466.3),
    ("EMEA", 38.1),
    ("APAC", 91.8)
]

for region_name, revenue in regional_2023_data:
    sales_rows.append({
        "year": 2023,
        "region": region_name,
        "channel": "Combined",
        "revenue_gbp": revenue,
        "gross_profit_gbp": None,
        "operating_profit_gbp": None,
        "pre_tax_profit_gbp": None,
        "net_profit_gbp": None,
        "notes": "Revenue from FashionNetwork Jun 3 2024."
    })

# -------------------------
# 2023 Channel Split
# -------------------------
channel_2023_data = [
    ("Retail", 238.9),
    ("Ecommerce", 134.9),
    ("Wholesale", 416.2),
    ("Outlet", 203.0)
]

for channel_name, revenue in channel_2023_data:
    sales_rows.append({
        "year": 2023,
        "region": "Global",
        "channel": channel_name,
        "revenue_gbp": revenue,
        "gross_profit_gbp": None,
        "operating_profit_gbp": None,
        "pre_tax_profit_gbp": None,
        "net_profit_gbp": None,
        "notes": "Channel revenue from FashionNetwork Jun 3 2024."
    })

# -------------------------
# 2024 Global (Combined)
# -------------------------
sales_rows.append({
    "year": 2024,
    "region": "Global",
    "channel": "Combined",
    "revenue_gbp": 901.3,
    "gross_profit_gbp": 432.0,
    "operating_profit_gbp": -18.1,
    "pre_tax_profit_gbp": -39.2,
    "net_profit_gbp": -39.3,
    "notes": "Public filing; FashionNetwork Jun 30 2025."
})

# -------------------------
# Placeholder Years (2019–2022)
# -------------------------
for y in [2019, 2020, 2021, 2022]:
    for region_name in ["UK/ROI", "Americas"]:
        sales_rows.append({
            "year": y,
            "region": region_name,
            "channel": "Combined",
            "revenue_gbp": None,
            "gross_profit_gbp": None,
            "operating_profit_gbp": None,
            "pre_tax_profit_gbp": None,
            "net_profit_gbp": None,
            "notes": "Placeholder – fill with internal or public data."
        })

# -------------------------
# Create DataFrame ONCE
# -------------------------
df_sales = pd.DataFrame(sales_rows, columns=columns_sales)

# Optional: enforce numeric dtypes
numeric_cols = [
    "revenue_gbp",
    "gross_profit_gbp",
    "operating_profit_gbp",
    "pre_tax_profit_gbp",
    "net_profit_gbp"
]

df_sales[numeric_cols] = df_sales[numeric_cols].astype("float64")
df_sales["year"] = df_sales["year"].astype("Int64")


# ============================================================
# 2) DEMOGRAPHICS DATA
# ============================================================

columns_demo = [
    "year",
    "region",
    "location_detail",
    "gender",
    "age_range",
    "income_level",
    "share_percent",
    "notes"
]

demo_rows = []

# UK 2023 – Men's 30–44 growth noted
demo_rows.append({
    "year": 2023,
    "region": "UK",
    "location_detail": None,
    "gender": "Male",
    "age_range": "30-44",
    "income_level": None,
    "share_percent": None,
    "notes": "Growth noted in public report; exact % not stated (FashionNetwork Jun 3 2024)."
})

# US 2023 – Average age drop
demo_rows.append({
    "year": 2023,
    "region": "US",
    "location_detail": None,
    "gender": None,
    "age_range": "Average ~41",
    "income_level": None,
    "share_percent": None,
    "notes": "Average age dropped from 49 pre-pandemic to 41 in 2023."
})

# US 2024 Placeholder
demo_rows.append({
    "year": 2024,
    "region": "US",
    "location_detail": None,
    "gender": None,
    "age_range": None,
    "income_level": None,
    "share_percent": None,
    "notes": "Placeholder – fill with survey or CRM data."
})

df_demo = pd.DataFrame(demo_rows, columns=columns_demo)

df_demo["year"] = df_demo["year"].astype("Int64")
df_demo["share_percent"] = df_demo["share_percent"].astype("float64")


# ============================================================
# 3) EXPORT TO CSV
# ============================================================

output_dir = Path("clarks_data_csv")
output_dir.mkdir(exist_ok=True)

sales_csv = output_dir / "clarks_sales_profit.csv"
demo_csv = output_dir / "clarks_demographics.csv"

df_sales.to_csv(sales_csv, index=False)
df_demo.to_csv(demo_csv, index=False)

print(f"Sales/profit CSV written to: {sales_csv.resolve()}")
print(f"Demographics CSV written to: {demo_csv.resolve()}")
```
