```python
import csv
import random

brands = ["Keen", "Cole Haan", "Saucony"]
years = [2022, 2023, 2024, 2025]
quarters = ["Q1", "Q2", "Q3", "Q4"]
regions = ["West", "South", "Northeast", "Midwest"]

# Example product categories per brand (simplified)
categories = {
    "Keen": ["Hiking Boots", "Sandals", "Trail Shoes", "Work Boots"],
    "Cole Haan": ["Dress Shoes", "Handbags", "Sneakers", "Outerwear"],
    "Saucony": ["Running Shoes", "Trail Running Shoes", "Performance Trainers"]
}

age_groups = ["18-24", "25-34", "35-44", "45-54"]
genders = ["Male", "Female"]
income_brackets = ["Under 50k", "50k-75k", "75k-100k", "100k-150k", "150k+"]
urbanicity = ["Urban", "Suburban", "Rural"]

random.seed(42)

rows = []
for brand in brands:
    for year in years:
        for q in quarters:
            for region in regions:
                prod = random.choice(categories[brand])
                units = random.randint(24000, 52000)
                # Example unit price range depending on brand/product; adjust as needed
                if brand == "Keen":
                    price = random.uniform(120, 150)  # avg unit revenue
                elif brand == "Cole Haan":
                    price = random.uniform(180, 250)
                else:  # Saucony
                    price = random.uniform(120, 160)
                revenue = round(units * price, 0)
                # COGS ratio
                cogs_ratio = random.uniform(0.58, 0.65)
                cogs = round(revenue * cogs_ratio, 0)
                gross_profit = revenue - cogs
                margin = round(gross_profit / revenue * 100, 1)
                age = random.choice(age_groups)
                gender = random.choice(genders)
                income = random.choice(income_brackets)
                urb = random.choice(urbanicity)

                rows.append([
                    brand, year, q, region, prod, units, revenue, cogs,
                    gross_profit, margin, age, gender, income, urb
                ])

# Write to CSV
with open("brand_sales_demo.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow([
        "Brand","Year","Quarter","Region","Product_Category","Units_Sold",
        "Revenue_USD","COGS_USD","Gross_Profit_USD","Profit_Margin_Percent",
        "Customer_Age_Group","Gender","Income_Bracket","Urbanicity"
    ])
    writer.writerows(rows)
```
