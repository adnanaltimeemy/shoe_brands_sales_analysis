```python
# -*- coding: utf-8 -*-
"""
Generate two CSV files with Dr. Martens sales, profit margin, and basic demographics
for UK and US contexts, using publicly reported data and traffic demographics.

Output files:
  - UK_data.csv
  - US_data.csv

Fields include:
  Year, Region, Revenue_GBP_m, Adjusted_EBIT_GBP_m, Adj_EBIT_Margin_pct,
  Comments, Web_Traffic_Share_pct, Demo_Age_Largest, Demo_Gender_Female_pct,
  Demo_Gender_Male_pct
"""

import csv

# ---------------------------------------------------------------------
# 1) Core data manually filled from cited public sources
#    FY2025 = year ended 30 March 2025
#    FY2024 = prior year
#
#    UK-specific revenue is cited as part of EMEA.
#    Americas revenue is used as proxy for US context with comment.
#
#    Adjusted EBIT values by region are reported for EMEA and Americas.
#
#    Source references are omitted here for brevity but were collected
#    from Dr. Martens public filings and Similarweb snapshots.
# ---------------------------------------------------------------------

# Data dictionary: keyed by (year, context)
# Values: dict of fields
data_rows = []

# ---------- UK context ----------
# FY2025 UK revenue (GBP million) reported as part of EMEA
#   UK revenue: £142.1m (FY25) ; FY24: £168.5m  - from footnote in segmental info
# FY2025 EMEA adjusted EBIT for reference, plus margin for EMEA region
#   Use EMEA as regional context for UK since UK is part of EMEA.
#   Adjusted EBIT: EMEA £77.3m (FY25) ; FY24: £109.7m
#   Adjusted EBIT margin for EMEA: 20.1% (FY25) ; 25.4% (FY24)  (as reported)
#
# Web traffic demographics: global site; note UK share of traffic ~12.45% (latest)
#   Gender: Female 63.51%; Male 36.49%; largest age group 25-34
#   These are global site metrics, not UK-only, so annotate.

data_rows.append({
    "Year": 2025,
    "Region": "UK",
    "Revenue_GBP_m": 142.1,               # UK revenue within EMEA
    "Adjusted_EBIT_GBP_m": "",            # Not separately disclosed for UK alone
    "Adj_EBIT_Margin_pct": "",            # Not separately disclosed for UK alone
    "Comments": ("UK-specific revenue published as part of EMEA. "
                 "EMEA adjusted EBIT reported; UK-only EBIT not provided."),
    "Web_Traffic_Share_pct": 12.45,       # latest share of traffic to drmartens.com from UK
    "Demo_Age_Largest": "25-34",
    "Demo_Gender_Female_pct": 63.51,
    "Demo_Gender_Male_pct": 36.49
})

data_rows.append({
    "Year": 2024,
    "Region": "UK",
    "Revenue_GBP_m": 168.5,               # UK revenue within EMEA FY24
    "Adjusted_EBIT_GBP_m": "",            # Not separately disclosed for UK alone
    "Adj_EBIT_Margin_pct": "",            # Not separately disclosed for UK alone
    "Comments": ("UK-specific revenue published as part of EMEA. "
                 "EMEA adjusted EBIT reported; UK-only EBIT not provided."),
    "Web_Traffic_Share_pct": 12.45,       # same snapshot used; historical UK-specific traffic not included
    "Demo_Age_Largest": "25-34",
    "Demo_Gender_Female_pct": 63.51,
    "Demo_Gender_Male_pct": 36.49
})

# ---------- US context ----------
# Use Americas figures as proxy, note limitation.
#
# FY2025 Americas revenue: £288.5m
# FY2024 Americas revenue: £325.8m
#
# Adjusted EBIT for Americas:
#   FY2025: £13.6m
#   FY2024: £41.7m
#
# Adjusted EBIT margin for Americas:
#   FY2025: 4.7%
#   FY2024: 12.8%   (reported adjusted margin by region)
#
# Comments: Americas includes USA; no US-only split provided.
data_rows.append({
    "Year": 2025,
    "Region": "US (Americas proxy)",
    "Revenue_GBP_m": 288.5,
    "Adjusted_EBIT_GBP_m": 13.6,
    "Adj_EBIT_Margin_pct": 4.7,
    "Comments": ("Americas region used as proxy for US; Americas includes "
                 "USA plus other markets. No US-only revenue/margin disclosed."),
    "Web_Traffic_Share_pct": 36.29,       # latest share of traffic to drmartens.com from US
    "Demo_Age_Largest": "25-34",
    "Demo_Gender_Female_pct": 63.51,
    "Demo_Gender_Male_pct": 36.49
})

data_rows.append({
    "Year": 2024,
    "Region": "US (Americas proxy)",
    "Revenue_GBP_m": 325.8,
    "Adjusted_EBIT_GBP_m": 41.7,
    "Adj_EBIT_Margin_pct": 12.8,
    "Comments": ("Americas region used as proxy for US; Americas includes "
                 "USA plus other markets. No US-only revenue/margin disclosed."),
    "Web_Traffic_Share_pct": 36.29,       # same snapshot; historical US-only traffic not available here
    "Demo_Age_Largest": "25-34",
    "Demo_Gender_Female_pct": 63.51,
    "Demo_Gender_Male_pct": 36.49
})

# ---------------------------------------------------------------------
# 2) Write separate CSVs for UK and US contexts
# ---------------------------------------------------------------------

# Shared field order
fieldnames = [
    "Year", "Region", "Revenue_GBP_m", "Adjusted_EBIT_GBP_m", "Adj_EBIT_Margin_pct",
    "Comments", "Web_Traffic_Share_pct", "Demo_Age_Largest",
    "Demo_Gender_Female_pct", "Demo_Gender_Male_pct"
]

# Filter rows for UK
uk_rows = [r for r in data_rows if r["Region"].startswith("UK")]
# Filter rows for US
us_rows = [r for r in data_rows if r["Region"].startswith("US")]

# Write UK CSV
with open("UK_data.csv", mode="w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for r in uk_rows:
        writer.writerow(r)

# Write US CSV
with open("US_data.csv", mode="w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for r in us_rows:
        writer.writerow(r)

print("UK_data.csv and US_data.csv have been generated.")
```
